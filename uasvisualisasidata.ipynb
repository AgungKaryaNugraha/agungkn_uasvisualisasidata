```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('kebutuhan_per_bulan_2.csv')

data_filtered_jabar = data[data['Propinsi'] == 'JAWA BARAT']

plt.scatter(data_filtered_jabar['tahun'], data_filtered_jabar['kebutuhan'], c=data_filtered_jabar.index, cmap='rainbow', s=50, alpha=0.7)

plt.title('Scatter Plot - Kebutuhan per Tahun (Jawa Barat)')
plt.xlabel('Tahun')
plt.ylabel('Kebutuhan')

cbar = plt.colorbar()
cbar.set_label('Data Index')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('SupplyKantor.csv')

data_tgl = data['TanggalPesan']
data_kota = data['Kota']
data_sales = data['SalesRep']
data_jenis = data['Jenis']
data_jum = data['Jumlah']
data_harga = data['Harga Satuan']

plt.scatter(data_tgl, data_harga, s=50, c='blue', alpha=0.7)

plt.title('Scatter Plot - Harga Satuan terhadap Tanggal Pesan')
plt.xlabel('Tanggal Pesan')
plt.ylabel('Harga Satuan')

plt.xticks(rotation=90)

plt.show()
```
